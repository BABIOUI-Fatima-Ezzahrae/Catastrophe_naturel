# ============= Nettoyage =============

```python
# Convertir la colonne 'date' en datetime
data2["date"] = pd.to_datetime(data2["date"], errors="coerce")
```

```python
# Remplir les valeurs manquantes par la médiane pour chaque code (par pays)
data2['tas'] = data2.groupby('code')['tas'].transform(
    lambda x: x.fillna(x.median())
)

# Remplir les valeurs restantes avec la médiane globale (fallback)
data2['tas'].fillna(data2['tas'].median(), inplace=True)
```

```python
# Vérification après nettoyage les valeurs null:
print("\n======= Missing values: =======\n",data2.isnull().sum().sort_values(ascending=False))
```

```python
# Convertir la colonne 'date' en datetime
data1["date"] = pd.to_datetime(data1["date"], errors="coerce")
```

```python
# Remplir les valeurs manquantes par la médiane pour chaque code (par pays)
data1['pr'] = data1.groupby('code')['pr'].transform(
    lambda x: x.fillna(x.median())
)

# Remplir les valeurs restantes avec la médiane globale (fallback)
data1['pr'].fillna(data1['pr'].median(), inplace=True)
```

```python
# Vérification après nettoyage les valeurs null:
print("\n======= Missing values: =======\n",data1.isnull().sum().sort_values(ascending=False))
```

```python
# Filtrer les données entre 2010 et 2024
start_date = "2010-01-01"
end_date = "2024-12-31"

data1 = data1[(data1["date"] >= start_date) & (data1["date"] <= end_date)]
data2 = data2[(data2["date"] >= start_date) & (data2["date"] <= end_date)]

# Vérification
print("\n======= Data1 date range =======\n", data1["date"].min(), "->", data1["date"].max())
print("\n======= Data2 date range =======\n", data2["date"].min(), "->", data2["date"].max())
```

```python
# Sauvegarde des données nettoyées dans un fichier CSV
folder_path = "../Data/Data_Cleanning"

os.makedirs(folder_path, exist_ok=True)

file_path1 = os.path.join(folder_path, "data_weather_pr.csv")
file_path2 = os.path.join(folder_path, "data_weather_tas.csv")

data1.to_csv(file_path1, index=False)
data2.to_csv(file_path2, index=False)
```
