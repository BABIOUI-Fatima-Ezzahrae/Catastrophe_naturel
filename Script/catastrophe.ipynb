============== Nettoyage : ==============

```python
# ========>> Types Modifier <<=========== #

data['Entry Date'] = pd.to_datetime(data['Entry Date'])
data['Last Update'] = pd.to_datetime(data['Last Update'])
data['Start Day'] = data['Start Day'].astype('Int64')
data['End Day'] = data['End Day'].astype('Int64')
```

```python
# Remplir les colonnes vides par 0 :
cols_zero = [
    "AID Contribution ('000 US$)",
    "No. Homeless",
    "Reconstruction Costs ('000 US$)",
    "Reconstruction Costs, Adjusted ('000 US$)"
]

for col in cols_zero:
    data[col] = data[col].fillna(0)
```

```python
# Remplir les colonnes vides par median :
cols_median = [
    "Total Deaths",
    "Total Affected",
    "No. Injured",
    "Insured Damage ('000 US$)",
    "Insured Damage, Adjusted ('000 US$)",
    "Total Damage ('000 US$)",
    "Total Damage, Adjusted ('000 US$)",
    "No. Affected"
]
for col in cols_median:
    data[col] = data[col].fillna(data[col].median())
```

```python
# Columns with missing info but we keep them
cols_high_nan = ["Magnitude", "Latitude", "Longitude"]
for col in cols_high_nan:
    data[col + "_missing"] = data[col].isnull().astype(int)
    data[col] = data[col].fillna(data[col].median())
```

```python
data["CPI"] = data["CPI"].fillna(data["CPI"].mean())
```

```python
# Création d'une colonne Start_Date complète à partir de l'année, du mois et du jour
data["Start_Date"] = pd.to_datetime(
    data["Start Year"].astype(str) + "-" +
    data["Start Month"].astype(str) + "-" +
    data["Start Day"].astype(str), 
    errors="coerce"  # les dates invalides seront transformées en NaT
)

# Création d'une colonne End_Date complète à partir de l'année, du mois et du jour
data["End_Date"] = pd.to_datetime(
    data["End Year"].astype(str) + "-" +
    data["End Month"].astype(str) + "-" +
    data["End Day"].astype(str), 
    errors="coerce"
)
```

```python

# ======= Drop invalid rows only =======
data = data.dropna(subset=["Start_Date", "End_Date"])
data = data[data["Start_Date"] <= data["End_Date"]]
```

```python
# Correction des cas illogiques
data.loc[data["Total Deaths"] > data["Total Affected"], "Total Deaths"] = \
    data["Total Affected"]
```

```python
# 🔹 Suppression des lignes où Start_Date est après End_Date (illogique)
data = data[data["Start_Date"] <= data["End_Date"]]
```

```python
# 🔹 Vérification
print("Nombre de lignes restantes :", data.shape[0])
print("Valeurs manquantes Start_Date :", data["Start_Date"].isnull().sum())
print("Valeurs manquantes End_Date :", data["End_Date"].isnull().sum())
print("Start_Date > End_Date :", (data["Start_Date"] > data["End_Date"]).sum())
```

```python
print("\n======= Missing values: =======\n",data.isnull().sum().sort_values(ascending=False))
```

```python
# Colonnes avec beaucoup de valeurs manquantes
cols_missing = [
    "River Basin",
    "Event Name",
    "Associated Types",
    "Origin",
    "External IDs",
    "GADM Admin Units",
    "Admin Units",
    "Location"
]

# Remplissage intelligent
for col in cols_missing:
    if col in ["River Basin", "Location"]:
        # On peut essayer de remplir par la valeur la plus fréquente par pays
        data[col] = data.groupby("Country")[col].transform(
            lambda x: x.fillna(x.mode().iloc[0] if not x.mode().empty else "Unknown")
        )
    elif col in ["Event Name", "Associated Types", "Origin", "External IDs"]:
        # Pour ces colonnes textuelles, si impossible de deviner, on met "Unknown"
        data[col] = data[col].fillna("Unknown")
    elif col in ["GADM Admin Units", "Admin Units"]:
        # Pour ces colonnes complexes/JSON, on peut mettre un dictionnaire vide si manquant
        data[col] = data[col].fillna("[]")
```

```python
print("\n======= Missing values: =======\n",data.isnull().sum().sort_values(ascending=False))
```

```python
# Sauvegarde des données nettoyées dans un fichier CSV
data.to_csv(file_path, index=False)

print("Cleaned dataset saved successfully.")
```
